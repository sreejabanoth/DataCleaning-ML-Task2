```python
import pandas as pd
import numpy as np
medical = pd.read_csv("/content/KaggleV2-May-2016.csv")
```

```python
medical.isnull().sum()
```

```python
medical = medical[medical['Age'] >= 0]
```

```python
medical['ScheduledDay'] = pd.to_datetime(medical['ScheduledDay'])
medical['AppointmentDay'] = pd.to_datetime(medical['AppointmentDay'])
```

```python
medical.to_csv("medical_cleaned.csv", index=False)
```
